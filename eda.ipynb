```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_path = "dataverse_files/Dataset"
```

```python
df_t1_acc_exp1 = pd.read_csv(data_path + "/T1/Accelerometer_Data/Expt_1.csv")
df_t1_ae_exp1 = pd.read_csv(data_path + "/T1/Acoustic_Emission_Data/Expt_1.csv")
df_t1_f_exp1 = pd.read_csv(data_path + "/T1/Force_Data/Expt_1.csv")

pd.concat([df_t1_acc_exp1.describe(), df_t1_ae_exp1.describe(), df_t1_f_exp1.describe()], axis=1)
```

```python
df_t1_acc_exp2 = pd.read_csv(data_path + "/T1/Accelerometer_Data/Expt_2.csv")
df_t1_ae_exp2 = pd.read_csv(data_path + "/T1/Acoustic_Emission_Data/Expt_2.csv")
df_t1_f_exp2 = pd.read_csv(data_path + "/T1/Force_Data/Expt_2.csv")

pd.concat([df_t1_acc_exp2.describe(), df_t1_ae_exp2.describe(), df_t1_f_exp2.describe()], axis=1)
```

### T1 Accelerometer data Per Exp

```python
fig, axs = plt.subplots(4, 2, figsize=(12,8))
# axs = axs.flatten()

for t in range(8):
    r = t % 4
    c = t // 4

    features = []
    for i in range(1, 13):
        df_temp = pd.read_csv(data_path + f"/T{t+1}/Accelerometer_Data/Expt_{i}.csv")
        features.append(df_temp.iloc[:, 0])
    axs[r, c].boxplot(features, boxprops=dict(color='blue'),
                            whiskerprops=dict(color='red'),
                            capprops=dict(color='green'),
                            medianprops=dict(color='yellow'),
                            flierprops=dict(markerfacecolor='purple', marker='o', markersize=5, linestyle='none'))
    axs[r][c].set_ylim(-10.0, 10.0)
    axs[r][c].set_xlabel('Data Type')
    axs[r][c].set_ylabel('Value')

#    
plt.subplots_adjust(wspace=0.4, hspace=0.6)

plt.tight_layout()
plt.show()
```
